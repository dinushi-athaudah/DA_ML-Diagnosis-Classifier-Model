```python
# ==============================
# Import Libraries
# ==============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Scikit-learn imports for modeling
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Handling imbalanced data
from imblearn.over_sampling import SMOTE

# Statistical test
from scipy.stats import shapiro

# Visualization purposes
from IPython.display import display, Markdown

# ==============================
# Load Dataset
# ==============================
df = pd.read_csv('Data/me_cfs_vs_depression_dataset.csv')

# Preview the first 5 rows
df.head()

# Print basic dataset info
print("Shape of the dataset:", df.shape)
print("\nData types of dataset:\n", df.dtypes)
print("\nNumber of duplicate rows:", df.duplicated().sum())
print("\nMissing values per column:\n", df.isnull().sum())

# ==============================
# Function: Check Normal Distribution
# ==============================
def is_normal_distribution(column):
    """
    Checks if a numeric column follows a normal distribution using Shapiro-Wilk test.
    Returns True if p-value > 0.05, else False.
    """
    # Shapiro-Wilk test requires at least 3 non-null values
    if len(column.dropna()) >= 3:
        stat, p_value = shapiro(column.dropna())
        return p_value > 0.05
    return False

# ==============================
# Function: Fill Missing Values
# ==============================
def fill_missing_values(df):
    """
    Handles missing values:
    - Numeric columns: Fill with mean if normally distributed, median otherwise
    - Categorical columns: Fill with mode
    """
    # Numeric columns
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    cols_with_missing = [col for col in numeric_cols if df[col].isnull().any()]
    
    for col in cols_with_missing:
        if is_normal_distribution(df[col]):
            df[col] = df[col].fillna(df[col].mean())
        else:
            df[col] = df[col].fillna(df[col].median())
    
    # Categorical columns
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df[col].isnull().any():
            df[col] = df[col].fillna(df[col].mode()[0])
    
    return df

# Apply missing value handling
df = fill_missing_values(df)

# Verify that missing values are handled
print("\nMissing values after handling:\n", df.isnull().sum())
```

```python
# ==============================
# Visualization & Outlier Handling
# ==============================

# ------------------------------
# Define features
# ------------------------------
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = [col for col in df.select_dtypes(include=['object', 'category']).columns if col != 'diagnosis']

# Combine all features for visualization
all_features = numerical_features + categorical_features

# ------------------------------
# Initial Visualization
# ------------------------------
display(Markdown("## 1. Initial Visualization of Features by Diagnosis"))

ncols = 3
nrows = math.ceil(len(all_features) / ncols)

fig, axes = plt.subplots(nrows, ncols, figsize=(16, 5*nrows))
axes = axes.flatten()

for i, feature in enumerate(all_features):
    if feature in numerical_features and df[feature].nunique() > 2:
        # Boxplot for numeric features
        sns.boxplot(
            data=df, 
            x='diagnosis', 
            y=feature, 
            ax=axes[i], 
            hue='diagnosis', 
            palette='Set2', 
            legend=False
        )
        axes[i].set_ylabel(feature.replace("_", " ").title())
    else:
        # Countplot for categorical features
        sns.countplot(
            data=df, 
            x=feature, 
            hue='diagnosis', 
            ax=axes[i], 
            palette='Set2'
        )
        axes[i].set_ylabel("Count")
    
    axes[i].set_title(f'{feature.replace("_", " ").title()} by Diagnosis')

plt.tight_layout()
plt.show()

# ------------------------------
# Outlier Removal (IQR Method)
# ------------------------------
def remove_outliers_columnwise(df, features):
    """
    Removes outliers for numeric columns using the IQR method.
    Only applies to columns with enough unique values (>5).
    """
    df_clean = df.copy()
    for col in features:
        if df_clean[col].nunique() > 5:  # skip features with few unique values
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean

# Apply outlier removal
df_clean = remove_outliers_columnwise(df, numerical_features)
print("Shape before outlier removal:", df.shape)
print("Shape after outlier removal:", df_clean.shape)

# ------------------------------
# Visualization AFTER Outlier Removal
# ------------------------------
display(Markdown("## 2. Visualization After Outlier Removal"))

fig, axes = plt.subplots(nrows, ncols, figsize=(16, 5*nrows))
axes = axes.flatten()

for i, feature in enumerate(all_features):
    if feature in numerical_features and df_clean[feature].nunique() > 2:
        sns.boxplot(
            data=df_clean,
            x='diagnosis',
            y=feature,
            ax=axes[i],
            hue='diagnosis',       # keep different colors
            palette='Set2',
            showfliers=False,      # hide extreme points
            dodge=False,
            legend=False
        )
        axes[i].set_ylabel(feature.replace("_", " ").title())
    else:
        sns.countplot(
            data=df_clean,
            x=feature,
            hue='diagnosis',
            ax=axes[i],
            palette='Set2'
        )
        axes[i].set_ylabel("Count")
    
    axes[i].set_title(f'{feature.replace("_", " ").title()} by Diagnosis (After Outlier Removal)')

# Remove empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# ------------------------------
# Correlation Heatmap
# ------------------------------
display(Markdown("## 3. Correlation Heatmap of Numerical Features"))

plt.figure(figsize=(8,6))
corr_after = df_clean[numerical_features].corr()  # correlation on cleaned data
sns.heatmap(corr_after, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Heatmap", fontsize=12)
plt.show()
```

```python
# ==============================
# 1. Feature Encoding
# ==============================
# Prepare the dataset for modeling by encoding categorical variables

df_model = df_clean.copy()  # use cleaned data after outlier removal

# --- Binary encoding ---
df_model['gender'] = df_model['gender'].map({'Male': 1, 'Female': 0})
df_model['meditation_or_mindfulness'] = df_model['meditation_or_mindfulness'].map({'Yes': 1, 'No': 0})

# --- Ordinal encoding ---
ordinal_mappings = {
    'social_activity_level': {'Very low': 0, 'Low': 1, 'Medium': 2, 'High': 3, 'Very high': 4},
    'exercise_frequency': {'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Often': 3, 'Daily': 4}
}
for col, mapping in ordinal_mappings.items():
    df_model[col] = df_model[col].map(mapping)

# --- One-hot encoding for nominal categorical feature ---
df_model = pd.get_dummies(df_model, columns=['work_status'], drop_first=True)

# --- Encode target variable ---
le_target = LabelEncoder()
df_model['diagnosis'] = le_target.fit_transform(df_model['diagnosis'])

# ==============================
# 2. Prepare Data for Modeling
# ==============================
X = df_model.drop(columns=['diagnosis'])
y = df_model['diagnosis']

# Show class distribution before balancing
print("\nTarget distribution before SMOTE:\n", y.value_counts())

# --- Handle class imbalance using SMOTE ---
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# --- Split dataset into training and testing sets ---
X_train, X_test, y_train, y_test = train_test_split(
    X_smote, y_smote, test_size=0.2, random_state=42, stratify=y_smote
)

# --- Scale features ---
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ==============================
# 3. Model Training & Evaluation
# ==============================
# Define models and hyperparameters for GridSearch
models_params = {
    'KNN': {
        'model': KNeighborsClassifier(),
        'params': {'n_neighbors': range(1,6), 'weights': ['uniform','distance']}
    },
    'Decision Tree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {'max_depth': range(1,6), 'min_samples_split':[2,4,6], 'min_samples_leaf':[1,2]}
    },
    'SVM': {
        'model': SVC(probability=True),
        'params': {'C':[0.1,1,10], 'kernel':['linear','rbf'], 'class_weight':[None,'balanced']}
    },
    'Naive Bayes': {
        'model': GaussianNB(),
        'params': {}
    },
    'Random Forest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {'n_estimators':[50,100], 'max_depth':[3,5], 'min_samples_split':[2,4]}
    }
}

# Dictionary to store model performance
results = {}

for model_name, mp in models_params.items():
    print(f"\n--- Training {model_name} ---")
    
    # Apply GridSearchCV if hyperparameters are provided
    if mp['params']:
        grid = GridSearchCV(mp['model'], mp['params'], cv=3, scoring='accuracy')
        grid.fit(X_train, y_train)
        model = grid.best_estimator_
        models_params[model_name]['best_estimator'] = model
        print(f"Best parameters for {model_name}: {grid.best_params_}")
    else:
        model = mp['model']
        model.fit(X_train, y_train)
        models_params[model_name]['best_estimator'] = model
    
    # Predictions
    y_pred = model.predict(X_test)
    
    # Store performance metrics
    classification_rep = classification_report(y_test, y_pred, output_dict=True)
    results[model_name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': classification_rep['weighted avg']['precision'],
        'Recall': classification_rep['weighted avg']['recall'],
        'F1 Score': classification_rep['weighted avg']['f1-score']
    }
    
    # Confusion matrix visualization
    plt.figure(figsize=(5,4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# ==============================
# 4. Model Performance Comparison
# ==============================
results_df = pd.DataFrame(results).T
print("\nModel Performance Comparison:\n")
display(results_df)
```

```python
# ==============================
# 5. Voting Ensemble Model (Clean Version)
# ==============================
# Evaluate all trained models and store model objects in the results
for model_name, mp in models_params.items():
    # Use the best estimator if GridSearchCV was applied, else the original model
    model = mp.get('best_estimator', mp['model'])
    results[model_name]['Model'] = model

# Convert results dictionary into DataFrame for easy sorting
results_df = pd.DataFrame(results).T

# Sort models by metrics
sorted_by_recall = results_df.sort_values(by='Recall', ascending=False)
sorted_by_precision = results_df.sort_values(by='Precision', ascending=False)
sorted_by_accuracy = results_df.sort_values(by='Accuracy', ascending=False)

# Function to select top models while avoiding duplicates
selected_models = set()

def get_best_model(sorted_df, selected_models):
    for model_name in sorted_df.index:
        if model_name not in selected_models:
            selected_models.add(model_name)
            return model_name

# Select top models based on different metrics
best_recall_model_name = get_best_model(sorted_by_recall, selected_models)
best_accuracy_model_name = get_best_model(sorted_by_accuracy, selected_models)
best_precision_model_name = get_best_model(sorted_by_precision, selected_models)

# Retrieve actual model objects
best_recall_model = results[best_recall_model_name]['Model']
best_accuracy_model = results[best_accuracy_model_name]['Model']
best_precision_model = results[best_precision_model_name]['Model']

# Print selected models for clarity
print(f"Best Model by Recall: {best_recall_model_name}")
print(f"Best Model by Accuracy: {best_accuracy_model_name}")
print(f"Best Model by Precision: {best_precision_model_name}")

# ------------------------------
# Create and train Voting Ensemble
# ------------------------------
voting_ensemble = VotingClassifier(
    estimators=[
        ('Best_Recall', best_recall_model),
        ('Best_Accuracy', best_accuracy_model),
        ('Best_Precision', best_precision_model)
    ],
    voting='soft'  # probability-based voting
)

# Train ensemble on training data
voting_ensemble.fit(X_train, y_train)

# Predictions on test set
y_pred_ensemble = voting_ensemble.predict(X_test)

# Evaluate ensemble
ensemble_accuracy = accuracy_score(y_test, y_pred_ensemble)
ensemble_classification_report = classification_report(y_test, y_pred_ensemble)

print(f"\nEnsemble Model Accuracy: {ensemble_accuracy}")
print(f"\nEnsemble Model Classification Report:\n{ensemble_classification_report}")

# Confusion matrix for ensemble
cm_ensemble = confusion_matrix(y_test, y_pred_ensemble)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_ensemble, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix: Voting Ensemble Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
